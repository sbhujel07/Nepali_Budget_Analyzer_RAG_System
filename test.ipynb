```python
import re

def clean_text(text):
    # remove page markers like --- Page 1 ---
    text = re.sub(r"--- Page \d+ ---", "", text)
    
    # remove standalone page numbers (lines with only numbers)
    text = re.sub(r"^\s*\d+\s*$", "", text, flags=re.MULTILINE)
    
    # remove OCR garbage like .१10.80४.110
    text = re.sub(r"\.\d*[०-९\d./]+", "", text)
    
    # remove multiple newlines
    text = re.sub(r"\n{2,}", "\n", text)
    
    # remove extra spaces
    text = re.sub(r"\s{2,}", " ", text)
    
    return text.strip()
```

```python
from pdf2image import convert_from_path
import pytesseract
import os

# -----------------------------
# 1. Setup paths
# -----------------------------

# Path to Tesseract OCR executable
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Path to Poppler bin folder
poppler_path = r"C:\Program Files\poppler-25.12.0\Library\bin"
```

```python

# PDF file path
pdf_path = r"Data\budget_speech_2081.pdf"

# Output text file
output_text_file = r"budget_speech_text.txt"


# 2. Convert PDF to images

try:
    images = convert_from_path(pdf_path, dpi=300, poppler_path=poppler_path)
    print(f"Number of pages found: {len(images)}")
except Exception as e:
    print("Error converting PDF to images:", e)
    exit()


# 3. Extract text from images

def extract_text_from_images(images):
    extracted_text = []
    for i, page in enumerate(images):
        text = pytesseract.image_to_string(page, lang="nep")
        text = clean_text(text)
        extracted_text.append({
            "page_number": i,
            "text": text
        })
    return extracted_text

text_data = extract_text_from_images(images)

# -----------------------------
# 4. Save text to file
# -----------------------------
with open(output_text_file, "w", encoding="utf-8") as f:
    for page in text_data:
        f.write(f"--- Page {page['page_number']} ---\n")
        f.write(page['text'])
        f.write("\n\n")

print(f"Extraction completed! Text saved to {output_text_file}")


# 5. Optional: print first page text

if text_data:
    print("\nFirst page preview:\n")
    print(text_data[0]["text"])
```

```python
import re

#define splitter  func
def split_nepali_sentences(text):
    abbreviations = ['रु', 'डा', 'श्री', 'नं', 'क्र', 'पृ']
    text = text.replace('\n', ' ')
    for abbr in abbreviations:
        text = text.replace(abbr + '.', abbr + '###')
    
    sentences = re.split(r'(?<=[।॥!?])\s+', text)
    sentences = [s.replace('###', '.') for s in sentences]
    sentences = [s.strip() for s in sentences 
                 if s.strip() and not re.fullmatch(r'[०-९]+\.', s.strip())]
    return sentences
```

For Chunking Strategy lets use the sliding Window.Yesma k hunxa vanda kati number of sentence ligdai janey and we also define a overlap jasma 1st chunking ko last sentence 2nd chunking ko first sentence hunxa yesley context ko semantic preserve garxa.

```python
def create_chunks(data,chunk_size=3,overlap=1):
    all_chunks = []

    steps = chunk_size-overlap
    for items in data:
        page_number = items["page_number"]
        sentence = items["Sentence"]


        #for each page and sentence
        for i in range(0,len(sentence),steps):
            chunks_sentence = sentence[i:i+chunk_size]
            chunks_sentence = "".join(chunks_sentence)
            all_chunks.append({"Page_number":page_number,
                               "Chunk_Text":chunks_sentence,
                               "Chunk_Count":len(chunks_sentence)})

    return all_chunks
```

Now lets Install FAISS Vector Database.It uses a ANN for the nearest neighbor search and retrieve the relevant text.

```python
import faiss 
import numpy as np

#lets suppose docs 
docs = [
     "काठमाडौं नेपालको राजधानी हो",
    "एभरेस्ट संसारको सबैभन्दा अग्लो हिमाल हो",
    "नेपालमा धेरै धार्मिक स्थलहरू छन्",
    "पशुपतिनाथ मन्दिर काठमाडौंमा छ",
]

#convert the docs to embeddings 
embedding = model.encode(docs).astype('float32') #usually numpy array 64 bit hunxa so faiss ley 32 bit ma matra kaam garxa
d = embedding.shape[1]

#create vector db in indices and vector
index = faiss.IndexFlatL2(d)
index.add(embedding)

query = "नेपालको सबैभन्दा अग्लो हिमाल कुन हो?"
query_vector = model.encode([query]).astype('float32')

distance,indices = index.search(query_vector,k=2)

print("Best chunks:")
for indice in indices[0]:
    print(f"{docs[indice]}")
```

```python
#use for good text 
import textwrap
def print_wrapped(text,wrap_length=80):
  wrapped_text = textwrap.fill(text,wrap_length)
  print(wrapped_text)
```

```python
d = text_chunks_embeddings.shape[1]

#create vector db
index = faiss.IndexFlatL2(d)
index.add(text_chunks_embeddings)

query = "शिक्षा मन्त्रालयको बजेट"
query_embedding = model.encode([query]).astype('float32')

distance,indices = index.search(query_embedding,k=2)

for indice in indices[0]: #indices is in 2d so takes indices[0]
    print(f"Chunk_Index: {indice}")
    print(f"Distance: {distance[0][list(indices[0]).index(indice)]:.4f}")
    print(f"Text: ")
    {print_wrapped(filtered_chunks[indice]["Chunk_Text"])}
    print("--"*25)
```
